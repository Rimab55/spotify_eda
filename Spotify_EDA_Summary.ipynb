# Summary


We took a bottom up approach to this analysis, starting with exploring what data was available through Spotify's API. This project combines track features available through the audio features endpoint with top tracks by country during the week of January 25th, 2019, and explores how scores across these features vary by country. Top tracks were scraped from the Spotify Charts web page. The aim of our analysis is to give insight into cultural differences as measured by musical taste profiles.
